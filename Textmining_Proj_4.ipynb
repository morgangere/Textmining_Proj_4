```python
# converting the text to a list.
reviews= df['review'].tolist()
```

```python
import nltk
#creating a fucntion to tokenize.
def tokenize(text):
    return nltk.word_tokenize(text)
```

```python
#selecting the stop words
nltkstopwords = nltk.corpus.stopwords.words('english')
stopwords = nltkstopwords
```

```python
# creating a function for the pipeline to remove stop words
def remove_stop(tokens):
    return [t for t in tokens if t not in stopwords]
```

```python
#selecting the punctuation to be removed
from string import punctuation
```

```python
# creating a function for the pipeline to remove punctuation
def remove_punc(tokens):
    return [t for t in tokens if t not in punctuation]
```

```python
## Porter stemming
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# creating a function for the pipeline to stem the tokens
def port_stem(tokens):
    return [ps.stem(t) for t in tokens]
```

```python
# creating a function to use each function from the pipeline on the texts
def prepare(text, pipeline):
    tokens = text
    for transform in pipeline:
        tokens = transform(tokens)
    return tokens                
```

```python
# selecting the functions for the pipeline
liepipeline = [str.lower,
            tokenize,
            remove_punc,
            remove_stop,
            port_stem
           ]
# using the pipelined list of functions.
lie_review_toke=[]
for review in reviews:
    x=prepare(review,liepipeline)
    lie_review_toke.append(x)
```

```python
# selecting the functions for the pipeline
sentpipeline = [str.lower,
            tokenize,
            remove_punc,
            remove_stop,
            #port_stem
           ]
# using the pipelined list of functions.
sent_review_toke=[]
for review in reviews:
    x=prepare(review,sentpipeline)
    sent_review_toke.append(x)
```

```python
#creating a new list
lie_review_for_vec=[]
#joining the lists of lists
#so a list of stirngs remains
for review in lie_review_toke:
    x=' '.join(review)
    #placing strings in new list
    lie_review_for_vec.append(x)
```

```python
#creating a new list
sent_review_for_vec=[]
#joining the lists of lists
#so a list of stirngs remains
for review in sent_review_toke:
    x=' '.join(review)
    #placing strings in new list
    sent_review_for_vec.append(x)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# selecing and saving the Tfid vectorizer with its properties
unigram_tfidf_vectorizer = TfidfVectorizer(encoding='latin-1', use_idf=True, min_df=2)
```

```python
#creating the lie vector
liereviewvec = unigram_tfidf_vectorizer.fit_transform(lie_review_for_vec)
```

```python
#creating a data frame from the lie vector.
pd.set_option('display.max_columns', None)
liedf=pd.DataFrame(liereviewvec.toarray(),
            columns=unigram_tfidf_vectorizer.get_feature_names_out())
```

```python
#creating the sentiment vector
sentreviewvec = unigram_tfidf_vectorizer.fit_transform(sent_review_for_vec)
```

```python
#creating a data frame from the sentiment vector.
pd.set_option('display.max_columns', None)
sentdf=pd.DataFrame(sentreviewvec.toarray(),
            columns=unigram_tfidf_vectorizer.get_feature_names_out())
```

```python
#converting the lie column
# from the original data frame to a list.
lie= df['lie'].tolist()
#creating an empty list.
lie_for_vec=[]
#transfroming t to 0
#transforming f to 1
for item in lie:
    if item == 't':
        item =0
    elif item =='f':
        item =1
    #placing the transformed variables in the new list.
    lie_for_vec.append(item)
print(lie_for_vec)
```

```python
#converting the sentiment column
# from the original data frame to a list.
sentiment= df['sentiment'].tolist()
#creating an empty list.
sent_for_vec=[]
#transfroming n to 0
#transforming p to 1
for item in sentiment:
    if item == 'n':
        item =0
    elif item =='p':
        item =1
    #placing the transformed variables in the new list.
    sent_for_vec.append(item)
print(sent_for_vec)
```

```python
#adding the lie labels 
#to the lie data frame
liedf['lie']=lie_for_vec
```

```python
#adding the sentiment labels 
#to the sentiment data frame
sentdf['sent']=sent_for_vec
```

```python
#selecting the features to predict from
sentfeature = sentdf[sentdf.columns[:-1]]
#selecting the labels.
sentlabel = sentdf['sent']
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import cross_val_score
#selecting the multinomial naive bayes classifier
mnb = MultinomialNB()
#creating the model
sentmodel=mnb.fit(sentfeature, sentlabel)
#using cross valadation to find accuracy.
sentscores = cross_val_score(mnb, sentfeature, sentlabel, cv=10)
sentscores
```

```python
#finding the overall average accuracy.
print(np.mean(sentscores))
```

```python
import numpy as np
from sklearn.inspection import permutation_importance
```
